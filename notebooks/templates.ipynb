## Basic Tactic: Reaching out to a Recruiter

**Why**: Recruiters are the gatekeepers for most positions. A follow-up can bring attention to your application, confirm its receipt, and inquire about next steps.

### Template
**Subject**: Follow-Up on {job_title} Application

Hi {recruiter_name}, I hope you’re doing well. I recently applied for the {job_title} position at {company} and wanted to follow up on my application. I believe my background in {specific skills/experience} aligns well with the role, and I’m excited about the opportunity to contribute to {company}. 

Please let me know if you need any additional information from me.  Thank you for considering my application. 

Best regards,

Joe Black
https://www.linkedin.com/in/joeblack949/
646.924.7718

***Attach Resume***

```python
template = """
Subject: Follow-Up on {job_title} Application

Hi {recruiter_name}, I hope you’re doing well. I recently applied for the {job_title} position at {company} and wanted to follow up on my application. I believe my background in {specific skills/experience} aligns well with the role, and I’m excited about the opportunity to contribute to {company}. 

Please let me know if you need any additional information from me.  Thank you for considering my application. 

Best regards,

Joe Black
https://www.linkedin.com/in/joeblack949/
646.924.7718

Attach Resume
"""

print(template.format(recruiter_name="", job_title="", company=""))
```

Joe Black
